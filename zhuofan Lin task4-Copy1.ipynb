```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.animation as animation

class Visualizer:
    def __init__(self):
        # Set up the figure and 3D axis
        self.fig = plt.figure()
        self.ax = self.fig.add_subplot(111, projection='3d')
        self.ax.set_xlim([-10, 10])
        self.ax.set_ylim([-10, 10])
        self.ax.set_zlim([-10, 10])

    def plot_particles(self, particles):
        """
        Plot the particles in 3D space.
        :param particles: List of particles with position and type attributes.
        """
        self.ax.clear()
        self.ax.set_xlim([-10, 10])
        self.ax.set_ylim([-10, 10])
        self.ax.set_zlim([-10, 10])

        for particle in particles:
            x, y, z = particle.position
            self.ax.scatter(x, y, z, label=particle.type, alpha=0.8)
        
        self.ax.legend()

    def animate(self, particles_generator, interval=50):
        """
        Animate particle movements.
        :param particles_generator: A generator yielding updated particles data.
        :param interval: Interval between frames in milliseconds.
        """
        def update(frame):
            particles = next(particles_generator)
            self.plot_particles(particles)
        
        ani = animation.FuncAnimation(self.fig, update, interval=interval, blit=False)
        plt.show()
```

```python
class Particle:
    def __init__(self, position, velocity, particle_type):
        self.position = np.array(position)
        self.velocity = np.array(velocity)
        self.type = particle_type

    def update_position(self, time_step=0.1):
        self.position += self.velocity * time_step
```

```python
def particle_simulation():
    # Initialize particles
    particles = [
        Particle([0, 0, 0], [0.1, 0.2, 0.1], "proton"),
        Particle([1, -1, 0], [-0.1, 0.1, 0.2], "electron"),
        Particle([-1, 1, -1], [0.2, -0.1, 0.1], "muon")
    ]
    
    while True:
        for particle in particles:
            particle.update_position()
        yield particles
```

```python
if __name__ == "__main__":
    visualizer = Visualizer()
    simulation_data = particle_simulation()
    visualizer.animate(simulation_data, interval=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt  # Importing matplotlib for visualization

# Define a simple Particle class
class Particle:
    def __init__(self, position):
        self.position = position

# Create a list of particles with random positions
num_particles = 100  # Example number of particles
particles = [Particle(np.random.uniform(-10, 10, size=3)) for _ in range(num_particles)]

# Accumulate heatmap data
heatmap = np.zeros((20, 20, 20))  # Example grid size
for particle in particles:
    grid_position = np.floor((particle.position + 10) / 1).astype(int)
    # Ensure grid_position is within bounds
    grid_position = np.clip(grid_position, 0, heatmap.shape[0] - 1)
    heatmap[tuple(grid_position)] += 1

# Create a figure and axes for visualization
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # Create a 3D axis

# Define a threshold for visualization
threshold = 0  # Set your desired threshold

# Overlay heatmap in your visualization
ax.voxels(heatmap > threshold, alpha=0.5)

plt.show()  # Display the plot
```
